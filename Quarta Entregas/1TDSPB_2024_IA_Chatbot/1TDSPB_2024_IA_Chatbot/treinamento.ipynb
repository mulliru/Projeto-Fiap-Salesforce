## Após identificarmos as colunas e alguns dados que temos em nosso arquivo csv vamos começar o treinamento da nossa ML

```python
X = df.drop('Possui alguma deficiência', axis = 1)
y = df['Possui alguma deficiência']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,y, test_size = 0.3, random_state = 10
)
```

```python
X_train.head()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
modelo = LinearRegression()
modelo.fit(X_train, y_train)
```

#### teste básico
