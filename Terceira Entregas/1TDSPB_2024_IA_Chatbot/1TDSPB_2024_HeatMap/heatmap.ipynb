```python
!pip install pandas matplotlib seaborn opencv-python

import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
import cv2 #opencv
```

#### Lendo o CSV (Planilha)  ###


```python
df = pd.read_csv("./Inicial/home.csv")
df.sample(3)
```

## Home 

- o primeiro heatmap a ser apresentado durante é a página inicial do nosso site. Este heatmap oferece uma visualização detalhada dos padrões de interação dos usuários com a página principal, revelando áreas de destaque, pontos de interesse e comportamentos de navegação.



```python
image = cv2.imread('./Inicial/home.jpeg')

image = cv2.cvtColor (image, cv2.COLOR_BGR2RGB)
         

plt.imshow(image)
df_escalado = df.copy()

df_escalado['X Position'] = df_escalado['X Position'] / 1.1
df_escalado['Y Position'] = df_escalado['Y Position'] / 0.9

sns.scatterplot(
    df_escalado,
    x = "X Position",
    y = "Y Position",
    c = "k", s=1
)

sns.kdeplot(
    df_escalado,
        x = "X Position",
        y = "Y Position",
        lw= 0.01, fill=True, alpha=.75, cmap='rocket'
)

#limitando tamanho das linhas
plt.xlim(0,image.shape[1])
plt.ylim(image.shape[0], 0)


plt.axis("off")
plt.tight_layout()
plt.savefig("heatmap.png", dpi = 2400)
plt.show()
```
